```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
pipeline_files = os.listdir('pipeline_data')
pipeline_overall_blended_accuracys = []
pipeline_overall_unblended_accuracys = []
pipeline_train_times = []
pipeline_bces=[]
pipeline_sem_bces = []
pipeline_unblendeds =[]
pipeline_blendeds = []
pipeline_cals = []
for i in pipeline_files:
    test_file = np.load('pipeline_data/'+str(i),allow_pickle=True)
    pipeline_unblendeds.append(test_file['unblended'])
    pipeline_blendeds.append(test_file['blended'])
    pipeline_overall_unblended_accuracys.append(test_file['overall_unblended_accuracy'])
    pipeline_overall_blended_accuracys.append(test_file['overall_blended_accuracy'])
    pipeline_train_times.append(test_file['train_time'])
    pipeline_bces.append(test_file['bce'])
    pipeline_sem_bces.append(test_file['sem_bce'])
    pipeline_cals.append(test_file['calibration'])
    
    

    
```

```python
custom_files = os.listdir('custom_data')
custom_overall_blended_accuracys = []
custom_overall_unblended_accuracys = []
custom_train_times = []
custom_bces=[]
custom_sem_bces = []
custom_unblendeds =[]
custom_blendeds = []
custom_cals = []
for i in custom_files:
    test_file = np.load('custom_data/'+str(i),allow_pickle=True)
    custom_unblendeds.append(test_file['unblended'])
    custom_blendeds.append(test_file['blended'])
    custom_overall_unblended_accuracys.append(test_file['overall_unblended_accuracy'])
    custom_overall_blended_accuracys.append(test_file['overall_blended_accuracy'])
    custom_train_times.append(test_file['train_time'])
    custom_bces.append(test_file['bce'])
    custom_sem_bces.append(test_file['sem_bce'])
    custom_cals.append(test_file['calibration'])
    
```

```python
variant_files = os.listdir('variant_data')
variant_overall_blended_accuracys = []
variant_overall_unblended_accuracys = []
variant_train_times = []
variant_bces=[]
variant_sem_bces = []
variant_unblendeds =[]
variant_blendeds = []
variant_cals = []
for i in variant_files:
    test_file = np.load('variant_data/'+str(i),allow_pickle=True)
    variant_unblendeds.append(test_file['unblended'])
    variant_blendeds.append(test_file['blended'])
    variant_overall_unblended_accuracys.append(test_file['overall_unblended_accuracy'])
    variant_overall_blended_accuracys.append(test_file['overall_blended_accuracy'])
    variant_train_times.append(test_file['train_time'])
    variant_bces.append(test_file['bce'])
    variant_sem_bces.append(test_file['sem_bce'])
    variant_cals.append(test_file['calibration'])
    
```

```python
pipeline_unblended_accs = []
pipeline_unblended_errs = []
for i in pipeline_unblendeds:
    pipeline_unblended_accs.append(i[1:,0].astype(float))
    pipeline_unblended_errs.append((i[1:,0].astype(float)*(1-i[1:,0].astype(float)))/i[1:,1].astype(float))
pipeline_unblended_accs = np.array(pipeline_unblended_accs)
pipeline_blended_accs = []
pipeline_blended_errs = []
for i in pipeline_blendeds:
    pipeline_blended_accs.append(i[1:,0].astype(float))
    pipeline_blended_errs.append((i[1:,0].astype(float)*(1-i[1:,0].astype(float)))/i[1:,1].astype(float))
pipeline_blended_accs = np.array(pipeline_blended_accs)

custom_unblended_accs = []
custom_unblended_errs = []
for i in custom_unblendeds:
    custom_unblended_accs.append(i[1:,0].astype(float))
    custom_unblended_errs.append((i[1:,0].astype(float)*(1-i[1:,0].astype(float)))/i[1:,1].astype(float))
custom_unblended_accs = np.array(custom_unblended_accs)
custom_blended_accs = []
custom_blended_errs = []
for i in custom_blendeds:
    custom_blended_accs.append(i[1:,0].astype(float))
    custom_blended_errs.append((i[1:,0].astype(float)*(1-i[1:,0].astype(float)))/i[1:,1].astype(float))
custom_blended_accs = np.array(custom_blended_accs)

variant_unblended_accs = []
variant_unblended_errs = []
for i in variant_unblendeds:
    variant_unblended_accs.append(i[1:,0].astype(float))
    variant_unblended_errs.append((i[1:,0].astype(float)*(1-i[1:,0].astype(float)))/i[1:,1].astype(float))
variant_unblended_accs = np.array(variant_unblended_accs)
variant_blended_accs = []
variant_blended_errs = []
for i in variant_blendeds:
    variant_blended_accs.append(i[1:,0].astype(float))
    variant_blended_errs.append((i[1:,0].astype(float)*(1-i[1:,0].astype(float)))/i[1:,1].astype(float))
variant_blended_accs = np.array(variant_blended_accs)
```

```python
x = [21.5,22.5,23.5,24.5,25.5,26.5]
```

```python
plt.errorbar(x, np.mean(pipeline_unblended_accs,axis=0).T,xerr=[.5,.5,.5,.5,.5,.5],yerr=np.sqrt(np.mean(pipeline_unblended_errs,axis=0)),label='pipeline')
plt.errorbar(x, np.mean(custom_unblended_accs,axis=0).T,xerr=[.5,.5,.5,.5,.5,.5],yerr=np.sqrt(np.mean(pipeline_unblended_errs,axis=0)),label='custom')
plt.errorbar(x, np.mean(variant_unblended_accs,axis=0).T,xerr=[.5,.5,.5,.5,.5,.5],yerr=np.sqrt(np.mean(pipeline_unblended_errs,axis=0)),label='variant')
plt.legend()
plt.title("Unblended Accuracies")
plt.xlabel("Brightest Mag")
plt.ylabel("Accuracy")
plt.show()
```
