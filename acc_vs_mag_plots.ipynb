```python
def plot_calibration(figsize=(10,5), fontsize=14,
                    footprint_type='pipeline', saveas=None, output_types=['pdf', 'png']):
    plt.figure(figsize=figsize)
    plt.plot((0,1), (0,1), color='black', label='Ideal')
    if footprint_type == 'pipeline':
        xerr_lo = np.array(p_median_gp_pipeline) - np.array(p_16percentile_gp_pipeline)
        xerr_hi = np.array(p_84percentile_gp_pipeline) - np.array(p_median_gp_pipeline)
        xerr = np.concatenate((xerr_lo.reshape(1,-1), xerr_hi.reshape(1,-1)), axis=0)
        plt.errorbar(x=p_median_gp_pipeline, y=calibration_gp_pipeline,
                    xerr=xerr, yerr=calibrationerr_gp_pipeline,
                    label='Gaussian process')
        xerr_lo = np.array(p_median_lr_pipeline) - np.array(p_16percentile_lr_pipeline)
        xerr_hi = np.array(p_84percentile_lr_pipeline) - np.array(p_median_lr_pipeline)
        xerr = np.concatenate((xerr_lo.reshape(1,-1), xerr_hi.reshape(1,-1)), axis=0)
        plt.errorbar(x=p_median_lr_pipeline, y=calibration_lr_pipeline,
                    xerr=xerr, yerr=calibrationerr_lr_pipeline,
                    label='Logistic regression')
        xerr_lo = np.array(p_median_cnn_pipeline) - np.array(p_16percentile_cnn_pipeline)
        xerr_hi = np.array(p_84percentile_cnn_pipeline) - np.array(p_median_cnn_pipeline)
        xerr = np.concatenate((xerr_lo.reshape(1,-1), xerr_hi.reshape(1,-1)), axis=0)
        plt.errorbar(x=p_median_cnn_pipeline, y=calibration_cnn_pipeline,
                    xerr=xerr, yerr=calibrationerr_cnn_pipeline,
                    label='CNN')
#     plt.title('Calibration')
    plt.legend(fontsize=fontsize)
    plt.ylabel('Blend fraction', fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    plt.xlabel('Model blend probability', fontsize=fontsize)
    plt.xticks(fontsize=fontsize)
    if saveas is not None:
        for extension in output_types:
            plt.savefig(saveas+'.'+extension)
    plt.show()
```

```python
'''
brightness_rank: A string or None. If not none, specify the rank of the galaxy
    being considered in a given blended footprint.
'''
def plot_blended_accuracy(brightness_rank=None, modelname='', figsize=(10,5), fontsize=14,
                          include=('pipeline', 'custom', 'variant'),
                          saveas=None, output_types=['pdf', 'png']):
    plt.figure(figsize=figsize)
    include_lowercase = {footprint_type.lower() for footprint_type in include}
    if 'pipeline' in include_lowercase:
        plt.errorbar(x=mag_centers, y=posacc_pipeline,
                     xerr=0.5*np.ones(len(mag_centers)), yerr=posaccerr_pipeline,
                     label='Pipeline')
    if 'custom' in include_lowercase:
        plt.errorbar(x=mag_centers, y=posacc_custom,
                     xerr=0.5*np.ones(len(mag_centers)), yerr=posaccerr_custom,
                     label='Custom')
    if 'variant' in include_lowercase:
        plt.errorbar(x=mag_centers, y=posacc_variant,
                     xerr=0.5*np.ones(len(mag_centers)), yerr=posaccerr_variant,
                     label='Variant')
    plt.title(modelname)
    if len(include) > 1:
        plt.legend(title='Footprint type', fontsize=fontsize)
    plt.ylabel('Blended accuracy', fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    xlabel = 'mag'
    if brightness_rank is not None:
        xlabel = xlabel + f'({brightness_rank})'
    plt.xlabel(xlabel, fontsize=fontsize)
    plt.xticks(fontsize=fontsize)
    if saveas is not None:
        for extension in output_types:
            plt.savefig(saveas+'.'+extension)
    plt.show()
```
